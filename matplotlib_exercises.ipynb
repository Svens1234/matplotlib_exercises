```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m=np.linspace(0,10,11)
```

```python
m
```

```python
c=3*10**8
```

```python
E=m*c**2
```

```python
E
```

```python
plt.plot(m,E,color='red',lw=5)
plt.title('E=mc^2')
plt.xlabel('M in g')
plt.ylabel('E in j')
plt.xlim(0,10)
```

```python
plt.plot(m,E,color='red',lw=5)
plt.title('E=mc^2')
plt.xlabel('M in g')
plt.ylabel('E in j')
plt.xlim(0,10)
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(m,E,color='red',lw=5)
```

```python
plt.plot(m,E,color='red',lw=5)
plt.title('E=mc^2')
plt.xlabel('M in g')
plt.ylabel('E in j')
plt.xlim(0,10)
plt.yscale("log")
plt.grid(which='both',axis='y')
```
